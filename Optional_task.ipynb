```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.title("Age")
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["daily_sleep_hours"])
plt.title("Daily Sleep Hours")
plt.xlabel("Daily Sleep Hours")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.histplot(df["phq9_score"])
plt.title("PHQ9 Score")
plt.xlabel("PHQ9 Score")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.histplot(df["cgpa"])
plt.title("CGPA")
plt.xlabel("CGPA")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python

plt.subplot(2,2,1)
sns.countplot(x="gender", data=df)
plt.title("Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.countplot(x="major_category", data=df)
plt.title("Major Category")
plt.xlabel("Major Category")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.countplot(x="has_part_time_job", data=df)
plt.title("Part Time Job")
plt.xlabel("Part Time Job")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.countplot(x="semester_type", data=df)
plt.title("Semester Type")
plt.xlabel("Semester Type")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

```python
print("Average CGPA:")
print(df["cgpa"].mean())
```

```python
print("Maximum CGPA:")
print(df["cgpa"].max())
```

```python
print("Minimum CGPA:")
print(df["cgpa"].min())
```

```python
print("Average CGPA by Part-Time Job Status:")
print(df.groupby("has_part_time_job")["cgpa"].mean())
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="has_part_time_job", y="cgpa", data=df)
plt.title("CGPA by Part-Time Job")

plt.subplot(1,2,2)
sns.boxplot(x="gender", y="cgpa", data=df)
plt.title("CGPA by Gender")

plt.tight_layout()
plt.show()
```

```python
# Analysis Summary

# 1. The average age of students is about 21 years.
#    The average CGPA of students is around 3.0.

# 2. Students with part-time jobs generally have lower CGPA
#    compared to students without part-time jobs.

# 3. The semester with the highest number of students is Spring.
#    Fall and Summer semesters have fewer students.

# 4. Interesting observations:
#    Female students show a wider spread in CGPA than male students.
#    Students without part-time jobs have more consistent CGPA values.
#    Some students study many hours but still have low CGPA,
#    which may be linked to stress or mental health factors.
```
