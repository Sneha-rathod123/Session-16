```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.title("Age")
plt.xlabel("Age")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.histplot(df["bmi"])
plt.title("BMI")
plt.xlabel("BMI")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.histplot(df["children"])
plt.title("Children")
plt.xlabel("Children")
plt.ylabel("Count")

plt.subplot(2,2,4)
sns.histplot(df["expenses"])
plt.title("Expenses")
plt.xlabel("Expenses")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

plt.subplot(2,2,1)
sns.countplot(x="sex", data=df)
plt.title("Sex")
plt.xlabel("Sex")
plt.ylabel("Count")

plt.subplot(2,2,2)
sns.countplot(x="smoker", data=df)
plt.title("Smoker")
plt.xlabel("Smoker")
plt.ylabel("Count")

plt.subplot(2,2,3)
sns.countplot(x="region", data=df)
plt.title("Region")
plt.xlabel("Region")
plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Expenses:")
print(df["expenses"].mean())
```

```python
print("Maximum Expenses:")
print(df["expenses"].max())

print("Minimum Expenses:")
print(df["expenses"].min())
```

```python
print("Average Expenses by Smoker:")
print(df.groupby("smoker")["expenses"].mean())
```

```python
plt.subplot(1,2,1)
sns.boxplot(x="smoker", y="expenses", data=df)
plt.title("Expenses by Smoker")

plt.subplot(1,2,2)
sns.boxplot(x="sex", y="expenses", data=df)
plt.title("Expenses by Sex")

plt.tight_layout()
plt.show()
```
